```python
movie_metadata_cleaned = data.loc[:,['movie_title', 'director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name',
             'num_critic_for_reviews', 'duration', 'genres', 'num_voted_users', 'plot_keywords', 'movie_imdb_link',
             'num_user_for_reviews', 'title_year','imdb_score']]
movie_metadata_cleaned
# movie_metadata_cleaned['movie_imdb_link']
```

```python
movie_metadata_cleaned['actor_1_name'] = data['actor_1_name'].replace(np.nan, 'unknown')
movie_metadata_cleaned['actor_2_name'] = data['actor_2_name'].replace(np.nan, 'unknown')
movie_metadata_cleaned['actor_3_name'] = data['actor_3_name'].replace(np.nan, 'unknown')
movie_metadata_cleaned['director_name'] = data['director_name'].replace(np.nan, 'unknown')
```

```python
movie_metadata_cleaned['genres'] = movie_metadata_cleaned['genres'].str.replace('|', ' ', regex=True)
```

```python
movie_metadata_cleaned['movie_title'] = movie_metadata_cleaned['movie_title'].str.lower()
```

```python
movie_metadata_cleaned['movie_title'][4]
```

```python
import unicodedata
```

```python
movie_metadata_cleaned['movie_title'] = movie_metadata_cleaned['movie_title'].apply(lambda x : unicodedata.normalize("NFKD",x).strip())
```
